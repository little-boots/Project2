# Database Initialization

In this notebook we will create a table inside a SQLite database called `db/killbill.sqlite` and populate it with the results of an [API call](https://arcos-api.ext.nile.works/v1/combined_county_annual?key=WaPo) to a DEA database. This is part of the second project of the *UNC Data Analytics Bootcamp*.

First, we will store in a DataFrame the results of an API call. Since the results are in JSON format, we will use the `read_json()` Pandas method.

```python
import pandas as pd

url = 'https://arcos-api.ext.nile.works/v1/combined_county_annual?key=WaPo'
df = pd.read_json(url)
```

Create csv files one per year plus one with the total to be used in D3 to test visualizations. 

```python
import numpy as np

df[df.year == 2006].count()
df.dropna(inplace=True)
```

```python
df['year'].value_counts()
```

```python
for year in range(2006, 2013):
    df[df.year == year].to_csv('static/data/combined_county_annual_' + str(year) + '.csv', index=False)
```
